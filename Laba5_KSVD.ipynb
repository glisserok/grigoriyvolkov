```python
import plotly.express as px
import pandas as pd
import numpy as np

iris = px.data.iris()

fig1_1 = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_length',
                      color='species', title='3D розподіл даних датасету Iris')
fig1_1.update_layout(title_x=0.5)
fig1_1.show()

fig1_2 = px.histogram(iris, x="petal_width", color="species", barmode='overlay',
                     title='Розподіл ширини пелюсток (Petal Width) за видами')
fig1_2.update_layout(title_x=0.5)
fig1_2.show()

iris_avg = iris.groupby(['species', 'sepal_width'])['sepal_length'].mean().reset_index()
fig1_3 = px.line(iris_avg, x='sepal_width', y='sepal_length', color='species',
                title='Зміна середнього Sepal Length залежно від Sepal Width')
fig1_3.update_layout(title_x=0.5)
fig1_3.show()

gapminder = px.data.gapminder()

gapminder['gdp'] = gapminder['gdpPercap'] * gapminder['pop']
df_2007 = gapminder[gapminder['year'] == 2007].nlargest(10, 'gdp')

fig2_1 = px.bar(df_2007, x='country', y='gdp', color='gdp', text='gdp',
               title='Топ-10 країн за рівнем ВВП у 2007 році',
               color_continuous_scale='Viridis')
fig2_1.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig2_1.update_layout(title_x=0.5, xaxis={'categoryorder':'total descending'})
fig2_1.show()

df_euro = gapminder[(gapminder['continent'] == 'Europe') & (gapminder['year'] == 2007)].nlargest(5, 'gdp')
fig2_2 = px.pie(df_euro, values='gdp', names='country',
               title='Розподіл ВВП серед топ-5 країн Європи (2007 рік)')
fig2_2.update_layout(title_x=0.5)
fig2_2.show()

df_dynamic = gapminder[(gapminder['continent'] == 'Oceania') | (gapminder['country'] == 'Canada')]
fig2_3 = px.bar(df_dynamic, x='year', y='pop', color='country', barmode='group',
               title='Динаміка зміни кількості населення (Канада та Океанія)')
fig2_3.update_layout(title_x=0.5)
fig2_3.show()

x1 = np.linspace(0, 2*np.pi, 200)
df_sin = pd.DataFrame({'x': x1, 'y': np.sin(x1)})
fig3_1 = px.line(df_sin, x='x', y='y', title='Графік функції y = sin(x)')
fig3_1.update_layout(title_x=0.5)
fig3_1.show()

x2 = np.linspace(-np.pi, np.pi, 200)
df_cos = pd.DataFrame({'x': x2, 'y': np.cos(x2)})
fig3_2 = px.line(df_cos, x='x', y='y', title='Графік функції y = cos(x)')
fig3_2.update_layout(title_x=0.5)
fig3_2.show()

x3 = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 200)
df_tan = pd.DataFrame({'x': x3, 'y': np.tan(x3)})
fig3_3 = px.line(df_tan, x='x', y='y', title='Графік функції y = tg(x)')
fig3_3.update_layout(title_x=0.5, yaxis_range=[-10, 10])
fig3_3.show()
```
