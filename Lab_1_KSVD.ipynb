```python

import numpy as np

vector_1 = np.arange(1, 21)
print("Завдання 1:", vector_1)

matrix_ones = np.ones((4, 4))
print("\nЗавдання 2:\n", matrix_ones)

matrix_rand_neg = np.random.uniform(-1, 1, (5, 5))
print("\nЗавдання 3:\n", matrix_rand_neg)

matrix_rand_pos = np.random.rand(3, 3)
print("\nЗавдання 4:\n", matrix_rand_pos)

vec_a = np.random.randint(1, 21, 6)
vec_b = np.random.randint(1, 21, 6)
print("\nЗавдання 5:")
print("Масив A:", vec_a)
print("Масив B:", vec_b)
print("Додавання:", vec_a + vec_b)
print("Віднімання:", vec_a - vec_b)
print("Множення:", vec_a * vec_b)

mat_a = np.random.randint(1, 6, (3, 2))
mat_b = np.random.randint(1, 6, (2, 4))
res_mult = np.dot(mat_a, mat_b)
print("\nЗавдання 6:\n", res_mult)

mat_to_transp = np.random.rand(3, 3)
print("\nЗавдання 7 (Оригінал):\n", mat_to_transp)
print("Транспонована:\n", mat_to_transp.T)

mat_diff1 = np.random.randint(1, 11, (3, 3))
mat_diff2 = np.random.randint(1, 11, (3, 3))
print("\nЗавдання 8 (Різниця):\n", mat_diff1 - mat_diff2)

mat_prod1 = np.random.randint(-5, 6, (3, 3))
mat_prod2 = np.random.randint(-5, 6, (3, 3))
print("\nЗавдання 9 (Добуток):\n", np.dot(mat_prod1, mat_prod2))

mat_sum = np.random.uniform(-1, 1, (3, 3))
row_sums = np.sum(mat_sum, axis=1).reshape(-1, 1)
print("\nЗавдання 10 (Матриця):\n", mat_sum)
print("Вектор-стовпчик сум:\n", row_sums)
```
